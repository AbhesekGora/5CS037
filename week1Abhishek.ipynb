```python

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

# 1
print("1. Addition:")
print("x + y =\n", x + y)
print()

# 2
print("2. Subtraction:")
print("x - y =\n", x - y)
print()

# 3
print("3. Multiplication with integers:")
print("x * 3 =\n", x * 3)
print("y * 2 =\n", y * 2)
print()

# 4.
print("4. Square of each element:")
print("x² =\n", x**2)
print("y² =\n", y**2)
print("v² =", v**2)
print()

# 5.
print("5. Dot products:")
print("v · w =", np.dot(v, w))
print("x · v =", np.dot(x, v))
print("x · y =\n", np.dot(x, y))
print()

# 6.
print("6. Concatenation:")
xyrow = np.concatenate((x, y), axis=0)
print("x and y along rows:\n", xyrow)

# Concatenate v and w along column
vwcol = np.column_stack((v, w))
print("v and w along columns:\n", vwcol)
```

```python

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])


# 1.
A_inv = np.linalg.inv(A)
print("A inverse =\n", A_inv)

eye = np.dot(A, A_inv)
print("A.A⁻¹ =\n", eye)
```

```python
# 2. Prove AB not equals to BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("AB =\n", AB)
print("BA =\n", BA)

are_equal = np.array_equal(AB, BA)
print("Are AB and BA equal?", are_equal)


```

```python
# 3.
AB=np.dot(A,B)
ABtransp = AB.T
print("AB transpose", ABtransp)

Btransp = B.T
Atransp = A.T
BT_AT = np.dot(Btransp, Atransp)
print("B transpose . A Transpose =\n", BT_AT)

equal = np.array_equal(ABtransp, BT_AT)
print("Are (AB)Transpose and BTranspose.ATranspose equal?", equal)
```
